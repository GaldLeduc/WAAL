```python
ffd_h=nn.Linear(13,1)
learn_r=0.01
criterion=nn.MSELoss()
opt_h = torch.optim.SGD(ffd_h.parameters(),lr=learn_r)

X_tr, Y_tr, X_te, Y_te = get_dataset('Boston2')

X_tr=X_tr[:100]
Y_tr=Y_tr[:100].view(100,1)

n_pool = len(Y_tr)
idxs_lb = np.zeros(n_pool, dtype=bool)
idxs_tmp = np.arange(n_pool)
np.random.shuffle(idxs_tmp)
idxs_lb[idxs_tmp[:10]] = True
idxs_ulb=np.arange(n_pool)[~idxs_lb]
print(idxs_lb)
print(idxs_ulb)
```

```python
def T1(theta,Xk,Y_pred,Nb,Nk):
    '''T1=torch.zeros(len(theta))
    for j in range(len(T1)):
        for i in range(len(Y_pred)):
            T1[j]+=Xk[i][j],theta[j]-Y_pred[i]/len(theta)
        T1=(1/(Nb+Nk)) * T1'''
    T1=0
    for i in range(len(Y_pred)):
        T1+=(torch.dot(Xk[i],theta)-Y_pred[i])**2
    T1=(1/(Nb+Nk)) * T1
    T1=T1.detach().numpy()[0]
    return T1

def change_leave(coordinate,theta,theta_iter,i=1):
    if(coordinate[-i]==3):
        i+=1
        coordinate,theta_iter=change_leave(coordinate,theta,theta_iter,i)
    else:
        coordinate[-i]+=1
        theta_iter[i]=theta[i]+(-1)**(coordinate[-i]-1)*0.01
        if(i!=1):
            for j in range(1,i):
                coordinate[-i+j]=1
                theta_iter[j]=theta[j]
    return coordinate,theta_iter

def test_tensor(tensor,theta,theta_iter,coordinate,X_tr,y_predicted,iters=0):
    test_theta=theta_iter
    sous_iter1=T1(test_theta,X_tr,y_predicted,2,len(y_predicted))
    test_theta[0]=test_theta[0]-0.01
    sous_iter2=T1(test_theta,X_tr,y_predicted,2,len(y_predicted))
    test_theta[0]=test_theta[0]+0.01
    sous_iter3=T1(test_theta,X_tr,y_predicted,2,len(y_predicted))
    sous_iter=[sous_iter1,sous_iter2,sous_iter3]
    idx=np.subtract(coordinate,np.ones(len(coordinate))).reshape(len(coordinate),1).astype(int)
    tensor[idx]=sous_iter
    if( 1 in coordinate or 2 in coordinate):
        iters+=1
        if(iters%100 == 0):
            print('iteration',iters)
        coordinate,theta_iter=change_leave(coordinate,theta,theta_iter,iters)
        tensor=test_tensor(tensor,theta,theta_iter,coordinate,X_tr,y_predicted)
    return tensor

y_true=Y_tr[idxs_lb]
y_predicted=ffd_h(X_tr[idxs_lb])#F.relu(ffd_h(X_tr[idxs_lb]))
loss=0
for i in range(len(y_true)):
    loss += (y_predicted[i][0]-y_true[i][0])*(y_predicted[i][0]-y_true[i][0])
loss = loss/len(y_true)
        
grad=torch.zeros(1,len(X_tr[0]))
theta=ffd_h.weight.data[0]
theta_iter=theta
size=np.ones(len(theta)-1,int)*3
tensor = np.ones(size,float)
origin=np.ones(len(theta),int)
values=test_tensor(tensor,theta,theta_iter,origin,X_tr[idxs_lb],y_predicted,0)
```
